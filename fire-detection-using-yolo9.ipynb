```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov9c.pt')
```

```python
model.train(
    data="/kaggle/input/fire-detection-using-yolo8/Attempt n2/data.yaml",  # Custom fire.yaml
    epochs=5,
    imgsz=640,
    batch=16,
    workers=2,
    device=0 )
```
